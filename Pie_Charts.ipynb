```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create plot
# Set size at firt line of code
plt.subplots(figsize=(8, 8))
# slices to hughlight
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# add the percentage of each wedge on the pie chart with autocpt, set explode values to tuple
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# slices to highlight
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# fig size
plt.subplots(figsize=(8, 8))
# add the percentage of each wedge on the pie chart with autocpt, set explode values to tuple, set colors to array
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Obj orient interafce
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# insert figsize within subplot
fig, ax = plt.subplots(figsize=(8, 8))
# values to pop 
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# autopct, add shadow, switch direction, add start angle, colors
ax.pie(y_axis,labels=x_axis, autopct='%.1f%%', shadow=True, explode=explode_values, startangle=85, counterclock=False, colors=colors)
plt.show()
```
