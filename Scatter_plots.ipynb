```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create plot using plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# create plot using plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# create plot using plt.plot()
# Change color and add data label
plt.scatter(y_axis, x_axis, c="red", label='Chicago')
# Chain gca() and invert_yaxis()
plt.gca().invert_yaxis()
# Add axis labels
plt.ylabel("Dates")
plt.xlabel("Fare($)")
plt.xlim(0,50)
# Add title
plt.title("Pyber Fare by Month")
#add legend
plt.legend()
plt.show()
```

```python
# Set markers equal to fare price
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust the size of markers by multiplying the data in the y-axis by any value.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
# Set marker size equal to new variable
plt.scatter(x_axis, y_axis, s=y_axis_larger)

```

```python
# FORMAT LIST COMP
# new_list = [expression for item in list if conditional]
```

```python
# list comp to format
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
# Change color to skyblue, and marker size * 5, add transparency, set line width of marker, switch axis
ax.scatter(y_axis, x_axis, color="skyblue", alpha=0.8, edgecolor="black", linewidth=2, s= [i * 5 for i in y_axis], label='Boston')
# invert y axis so jan is at top
plt.gca().invert_yaxis()
# add labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
# add title
plt.title("Pyber Fare by Month")
# Add Limit
plt.xlim(0,50)
```
